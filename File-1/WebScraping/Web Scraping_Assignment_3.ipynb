```python
    
driver=webdriver.Chrome()               # for first connecting the google chrome webdriver
driver.get("https://www.amazon.in")     # for opening the website(https://www.amazon.in) in google chrome webdriver

# first click in search field
First_click= driver.find_element(By.XPATH,'//input[@class="nav-input nav-progressive-attribute"]')
First_click.click()


#Asking the user for input to search    
Searchproduct_input=input("Enter any product to search box in amazon:")

    
#To find element for search bar and Enter the product which is taken by user
search=driver.find_element(By.ID,"twotabsearchtextbox")
search.send_keys(Searchproduct_input)
    
#To find element for Search button and click on the search button
Search_button=driver.find_element(By.XPATH,'//input[@type="submit"]')
Search_button.click()

```

# Question 2

In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
#Fetching Product URL
Product_Urls=[]
# giving the start and end range of the pages for scraping the first 3 page's urls of product
start=0  # start page
end=3   # end page
for Page in range(start,end):
    #Scraping all the product Urls with the iteration process from start to end page
    for i in driver.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-4"]/a'):
        Product_Urls.append(i.get_attribute('href'))
    #clicking on the next button for jumping from one to another page (if we use .find_elements, then no need to write nextbutton.click())
    nextbutton=driver.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[68]/div/div/span/a[3]')


#for printing the len of all the scraped product urls   
len(Product_Urls)
```

```python

Brand_Name=[]
Product_Name=[]
Product_Price=[]
Return_Exchange=[]
Expected_Delivery=[]
Product_Availability=[]
#v=0 # used the variable v for knowing that which number of product details we are scraping

for i in Product_Urls:# To iterate for getting the details("Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability") of product one by one
    #print(v)
    driver.get(i) #for opening the product webpage
    time.sleep(1)  # for delay of 1s
    #v=v+1
    # scraping the details("Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability") of product
    # if any of the details are missing for any of the product then replace it by “-“ by using NoSuchElementException
    try: 
        brandname=driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]/td[2]/span')
        Brand_Name.append(brandname.text)
    except NoSuchElementException:
        Brand_Name.append("-")
    try:
        productname=driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Product_Name.append(productname.text)
    except NoSuchElementException:
        Product_Name.append("-")
    try:
        price=driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center"]/span[2]/span[2]')
        Product_Price.append(price.text.split("\n")[0])
    except NoSuchElementException:
        Product_Price.append("-")
    try:
        returnExchange=driver.find_element(By.XPATH,'//li[@class="a-carousel-card tw-scroll-carousel-element"]//div[1]/span[1]/div[2]/span[1]')
        Return_Exchange.append(returnExchange.text)
    except NoSuchElementException:
        Return_Exchange.append("-")
    try:
        expectedDelivery=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none"]/div[1]/div[1]/span[1]')
        Expected_Delivery.append(expectedDelivery.text.split("\n")[0-1])
    except NoSuchElementException:
        Expected_Delivery.append("-")
    try:
        availability=driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-micro }"]/span[1]')
        Product_Availability.append(availability.text)
    except NoSuchElementException:
        Product_Availability.append("-")
        
```

```python
# getting the length of all product details(Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL)
print(len(Brand_Name),len(Product_Name),len(Product_Price),len(Return_Exchange),len(Expected_Delivery),len(Product_Availability),len(Product_Urls))
```

```python
# printing all product details(Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL)
print(Brand_Name,Product_Name,Product_Price,Return_Exchange,Expected_Delivery,Product_Availability,Product_Urls)
```

```python
#creating a dataframe of the scraped data
df=pd.DataFrame({'Brand Name':Brand_Name,'Product Name':Product_Name,'Price':Product_Price,'Return/Exchange':Return_Exchange,'Expected Delivery':Expected_Delivery,'Availability':Product_Availability, 'Product_url':Product_Urls}) 
print(df.shape)
print(df)

```

```python
# Save DataFrame to CSV
df.to_csv("amazon_products.csv", index=False)

```

```python
Brand_Name=[]
Smartphone_Name=[]
Product_color=[]
Product_RAM=[]
Product_ROM=[]
Product_PrimaryCamera=[]
Product_SecondaryCamera=[]
Display_Size=[]
Battery_Capacity=[]
Product_Price=[]
#v=0
#scraping the details(“Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, 
#“Display Size”, “Battery Capacity”, “Price”, “Product URL) one by one after opening every scraped url of product
for i in Product_Urls: 
    #print(v)
    driver.get(i)
    time.sleep(1)
    #v=v+1
    try:
        brandname=driver.find_element(By.XPATH,'//div[@class="_1MR4o5"]/div[4]/a')
        Brand_Name.append(brandname.text.replace('Mobiles',''))
    except NoSuchElementException:
        Brand_Name.append("-")
    try:
        smartphonename=driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        Smartphone_Name.append(smartphonename.text)
        #Smartphone_Name.append(smartphonename.text.strip('()'))
    except NoSuchElementException:
        Smartphone_Name.append("-")
    try:
        color=driver.find_element(By.XPATH,'//table[@class="_14cfVK"]/tbody/tr[4]/td[2]/ul/li')
        Product_color.append(color.text)
    except NoSuchElementException:
        Product_color.append("-")
    try:
        RAM=driver.find_element(By.XPATH,"//ul/li[contains(text(),'RAM')]")
        Product_RAM.append(RAM.text.split("|")[0])
    except NoSuchElementException:
        Product_RAM.append("-")
    try:
        ROM=driver.find_element(By.XPATH,"//ul/li[contains(text(),'ROM')]")
        Product_ROM.append(ROM.text.split("|")[1])
    except NoSuchElementException:
        Product_ROM.append("-")
    try:
        PrimaryCamera=driver.find_element(By.XPATH,"//ul/li[contains(text(),'Rear Camera')]")
        Product_PrimaryCamera.append(PrimaryCamera.text.split("|")[0])
    except NoSuchElementException:
        Product_PrimaryCamera.append("-")
    try:
        SecondaryCamera=driver.find_element(By.XPATH,"//ul/li[contains(text(),'Front Camera')]")
        Product_SecondaryCamera.append(SecondaryCamera.text.split("|")[1])
    except NoSuchElementException:
        Product_SecondaryCamera.append("-")
    try:
        dispalysize=driver.find_element(By.XPATH,"//ul/li[contains(text(),'Display')]")
        Display_Size.append(dispalysize.text)
    except NoSuchElementException:
        Display_Size.append("-")
    try:
        batteryCapacity=driver.find_element(By.XPATH,"//ul/li[contains(text(),'Battery')]")
        Battery_Capacity.append(batteryCapacity.text)
    except NoSuchElementException:
        Battery_Capacity.append("-")
    try:
        Price=driver.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
        Product_Price.append(Price.text)
    except NoSuchElementException:
        Product_Price.append("-")
        
```

Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.    

```python
#importing the required libraries
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver=webdriver.Chrome()        # for just opening the google chrome webdriver
driver.get("https://www.youtube.com/watch?v=Ez8F0nW6S-w&t=793s")   # for opening the youtube in google chrome webdriver
driver.maximize_window() # maximise the window
```

```python
comments=[]
comment_Upvotes=[]
comment_time=[]
# scraping the details of the youtube video from starting to 20 times scrolling the webpage
for _ in range(20):
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(2) # delay for loading the comments
    
    for i in driver.find_elements(By.ID,"content-text"):
        comments.append(i.text.replace("\n"," "))
        
            
    for j in driver.find_elements(By.ID,"vote-count-middle"):
        comment_Upvotes.append(j.text)
       
            
    for k in driver.find_elements(By.XPATH,"//yt-formatted-string/a[contains(@href,'/watch?v=Ez8F0nW6S-w&lc')]"):
        comment_time.append(k.text)
        
```

```python
# printing the length of the details of Comments, Comment upvote and time 
print(len(comments),len(comment_Upvotes),len(comment_time))
```

Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver=webdriver.Chrome()        # for just opening the google chrome webdriver
driver.get("https://www.hostelworld.com/")   # for opening the https://www.hostelworld.com/ in google chrome webdriver
driver.maximize_window()             # for maximize the window
```

```python

#To find element for location bar and Enter London
Location= driver.find_element(By.XPATH,'//input[@class="native-input"]')
Location.send_keys("London")
time.sleep(2)
# Press Enter to confirm the input
Location.send_keys(Keys.ENTER)
time.sleep(2) # Wait for the scrollable dropdown to appear

```

```python
# Scroll to the desired item in the dropdown and click it
desired_item = driver.find_element(By.XPATH, '//li[contains(@class,"item is-two-row")][1]')  # Change index as needed
driver.execute_script("arguments[0].scrollIntoView();", desired_item)
desired_item.click()
```

```python
#To find element for Let's go button and click on Let's go button
Search_button=driver.find_element(By.XPATH,'//button[@class="btn-content large-button"]/div')
Search_button.click()
```

```python
Hostel_Name=[]
Distance=[]
Ratings=[]
Privates_price=[]
Dorms_price=[]
Total_Reviews=[]
Overall_Reviews=[]
for i in driver.find_elements(By.XPATH,'//div[@class="property-name"]/span'):
    Hostel_Name.append(i.text)
for i in driver.find_elements(By.XPATH,'//div[@class="property-distance"]'):
    Distance.append(i.text.replace("\n",""))
for i in driver.find_elements(By.XPATH,'//span[@class="number"]'):
    Ratings.append(i.text)
for i in driver.find_elements(By.XPATH,'//div[@class="property-accommodations"]/div[1]/div/div[3]/strong'):
    Privates_price.append(i.text)
for i in driver.find_elements(By.XPATH,'//div[@class="property-accommodations"]/div[1]/div[2]/div/strong'):
    Dorms_price.append(i.text)
for i in driver.find_elements(By.XPATH,'//div[@class="review"]/span'):
    Total_Reviews.append(i.text.strip('()'))
for i in driver.find_elements(By.XPATH,'//span[@class="keyword"]'):
    Overall_Reviews.append(i.text)
    
```
