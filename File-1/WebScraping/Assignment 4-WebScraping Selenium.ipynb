# Question 1

1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver=webdriver.Chrome()               # for first connecting the google chrome webdriver
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")     # for opening the website(https://www.amazon.in) in google chrome webdriver
```

```python
Rank=[]
Name=[]
Uploader=[]
Date=[]
Views=[]

for i in range(1,31):
    Rank.append(i)
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]'):
        Name.append(i.text.replace('"', '').split('[')[0])
except NoSuchElementException:
    Name.append("-")
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]'):
        Uploader.append(i.text)
except NoSuchElementException:
    Uploader.append("-")        
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]'):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("-")
try:
    for i in driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]'):
        Date.append(i.text)
except NoSuchElementException:
    Date.append("-")
```

```python
#printing the length of the scraped details
print(len(Rank),len(Name),len(Uploader),len(Views),len(Date))
#printing the scraped details
print(Rank,Name,Uploader,Views,Date)
```

```python
#creating a dataframe of the scraped data
df=pd.DataFrame({'Rank':Rank,'Name':Name,'Artist':Uploader,'Views (billions)':Views,'Upload Date':Date}) 
print(df.shape)
df
```

```python
Repository_URL=[]
for Page in range(0,1):
    for i in driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a'):
        Repository_URL.append(i.get_attribute('href'))
    
    
len(Repository_URL)
```

```python
Repository_Title=[]
RepositoryDescription=[]
Language_Used=[]
try:
    for i in driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a'):
        Repository_Title.append(i.text)
except NoSuchElementException:
    Repository_Title.append("-")
try:
    for i in driver.find_elements(By.XPATH,'//article[@class="Box-row"]/p'):
        RepositoryDescription.append(i.text)
except NoSuchElementException:
    RepositoryDescription.append("-")

try:
    for i in driver.find_elements(By.XPATH,'//article[@class="Box-row"]/div[2]/span[1]/span[2]'):
        Language_Used.append(i.text)
except NoSuchElementException:
    Language_Used.append("-")
```

```python
# Open Chrome browser
driver=webdriver.Chrome()  
driver.maximize_window()

# Navigate to billboard.com
driver.get("https://www.billboard.com/")
time.sleep(2)

# Click on the Charts option
click_charts = driver.find_element(By.XPATH, '//div[@id="main-wrapper"]/header/div/div[2]/div/div/div[2]/div[2]/div/div/nav/ul/li[1]/a')
click_charts.click()
time.sleep(1)

# Click on the Hot 100 page link
hot_100_link = driver.find_element(By.XPATH, '//li[@class="o-nav__list-item "][1]/a')
hot_100_link.click()
time.sleep(2)

Song_Name=[]
Artist_Name=[]
Last_Week_Rank=[]
Peak_Rank=[]
Weeks_On_Chart=[]
try:
    for i in driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li/h3'):
        Song_Name.append(i.text)
except NoSuchElementException:
    Song_Name.append("-")
try:
    for i in driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[1]/span'):
        Artist_Name.append(i.text)
except NoSuchElementException:
    Artist_Name.append("-")
try:
    for i in driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[4]/span'):
        Last_Week_Rank.append(i.text)
except NoSuchElementException:
    Last_Week_Rank.append("-")
try:
    for i in driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[5]/span'):
        Peak_Rank.append(i.text)
except NoSuchElementException:
    Peak_Rank.append("-")
try:
    for i in driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[6]/span'):
        Weeks_On_Chart.append(i.text)
except NoSuchElementException:
    Weeks_On_Chart.append("-")
    
print(len(Song_Name),len(Artist_Name),len(Last_Week_Rank),len(Peak_Rank),len(Weeks_On_Chart))

driver.close()

#creating a dataframe of the scraped data
df=pd.DataFrame({'Song Name':Song_Name,'Artist Name':Artist_Name,'Last Week Rank':Last_Week_Rank,'Peak_Rank':Peak_Rank, 'Weeks On Chart':Weeks_On_Chart}) 
print(df.shape)
df
```

# Question 8

Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/ You
have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year
Note: - from the home page you have to go to the Show All Dataset page through code.

```python
from selenium.common.exceptions import ElementClickInterceptedException

# Open Chrome browser
driver=webdriver.Chrome()  
driver.maximize_window()

# Navigate to billboard.com
driver.get("https://archive.ics.uci.edu/")
time.sleep(2)

# Click on the Charts option
click_datasets = driver.find_element(By.XPATH, '//a[@class="btn-primary btn"]')
click_datasets.click()
time.sleep(1)

Dataset_URL=[]
start=0
end =67
for Page in range(start,end):
    for i in driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a'):
        Dataset_URL.append(i.get_attribute('href'))
    for _ in range(10):
        driver.execute_script("window.scrollBy(0,500)")
    if(Page!=66):
        try:
            next_button=driver.find_element(By.XPATH,'//div[@class="btn-group"]/button[2]')
            next_button.click()
        except ElementClickInterceptedException:
            print("completed")
    time.sleep(3)
        
    
```

```python
len(Dataset_URL)
```

```python
Dataset_Name=[]
Data_Type=[]
Task=[]
Attribute_Type=[]
Number_Of_Instances=[]
Number_Of_Attributes=[]
datasetYear=[]

for i in Dataset_URL:
    driver.get(i)
    time.sleep(3)
    try:
        for i in driver.find_elements(By.XPATH,'//main[@class="min-h-screen"]/div[1]/div[1]/div[1]/div[1]/div[2]/div/h1'):
            Dataset_Name.append(i.text)
    except NoSuchElementException:
        Dataset_Name.append("-")
    try:
        for i in driver.find_elements(By.XPATH,'//main[@class="min-h-screen"]/div[1]/div[1]/div[1]/div[2]/div[2]/div[1]/p'):
            Data_Type.append(i.text)
    except NoSuchElementException:
        Data_Type.append("-")
    try:
        for i in driver.find_elements(By.XPATH,'//main[@class="min-h-screen"]/div[1]/div[1]/div[1]/div[2]/div[2]/div[3]/p'):
            Task.append(i.text)
    except NoSuchElementException:
        Task.append("-")
    try:
        for i in driver.find_elements(By.XPATH,'//main[@class="min-h-screen"]/div[1]/div[1]/div[1]/div[2]/div[2]/div[4]/p'):
            Attribute_Type.append(i.text)
    except NoSuchElementException:
        Attribute_Type.append("-")
    try:
        for i in driver.find_elements(By.XPATH,'//main[@class="min-h-screen"]/div[1]/div[1]/div[1]/div[2]/div[2]/div[5]/p'):
            Number_Of_Instances.append(i.text)
    except NoSuchElementException:
        Number_Of_Instances.append("-")
    try:
        for i in driver.find_elements(By.XPATH,'//main[@class="min-h-screen"]/div[1]/div[1]/div[1]/div[2]/div[2]/div[6]/p'):
            Number_Of_Attributes.append(i.text)
    except NoSuchElementException:
        Number_Of_Attributes.append("-")
    try:
        for i in driver.find_elements(By.XPATH,'//main[@class="min-h-screen"]/div[1]/div[1]/div[1]/div[1]/div[2]/h2'):
            datasetYear.append(i.text.split()[2])
    except NoSuchElementException:
        datasetYear.append("-")
        
```

```python
Years=[i.split("/")[-1] for i in datasetYear]
```

```python
print(len(Dataset_Name),len(Data_Type),len(Task),len(Attribute_Type),len(Number_Of_Instances),len(Number_Of_Attributes),len(Years))
```

```python
print(Dataset_Name,Data_Type,Task,Attribute_Type,Number_Of_Instances,Number_Of_Attributes,Years)
```
